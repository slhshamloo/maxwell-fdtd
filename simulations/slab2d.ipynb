# 2d Slabs

## Imports

```python
import slhfdtd
from matplotlib import pyplot as plt
```

## Global Configurations

```python
plt.rcParams.update({"text.usetex" : True})
```

## Simulation

```python
solver = slhfdtd.Solver((6e-6, 6e-6, 0), 3e-8)
solver.add_source(slhfdtd.PointSource((3e-6, 3e-6, 0), wavelength=400e-9))
solver.add_object(slhfdtd.Slab((1e-6, 2e-6, 0), (2e-6, 4e-6, 0), permittivity=2))
solver.add_object(slhfdtd.Slab((4e-6, 2e-6, 0), (5e-6, 4e-6, 0), permeability=2))
solver.add_object(slhfdtd.Slab((2.1e-6, 4.1e-6, 0), (3.9e-6, 5.1e-6, 0), conductivity=1))
solver.add_boundary(slhfdtd.AutoPML())
solver.run(2e-14)

visualizer = slhfdtd.Visualizer(solver)
visualizer.set_aspect('equal')
visualizer.set_interpolation_2d('none')
visualizer.set_variables(norms='log')
visualizer.plot2d_magnitude()

# plt.savefig('slab2d.pdf', bbox_inches='tight')
```
