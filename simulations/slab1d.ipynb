# 1D Slabs

## Imports

```python
import slhfdtd
from matplotlib import pyplot as plt
```

## Global Configurations

```python
plt.rcParams.update({"text.usetex" : True})
```

## Non-Conducting

```python
solver = slhfdtd.Solver((2e-6, 0, 0), grid_dist=1e-9)
solver.add_source(slhfdtd.PointSource((2e-9, 0, 0)))
solver.add_object(slhfdtd.Slab((1e-6, 0, 0), (1.3e-6, 0, 0), permittivity=10))
solver.add_boundary(slhfdtd.Exact1DAbsorber(0))
solver.run(2e-14)

fig, axs = plt.subplots(2)
visualizer = slhfdtd.Visualizer(solver)
visualizer.set_colors(object_colors='lightskyblue')
visualizer.plot1d_field(axs[0], 'E')
visualizer.plot1d_field(axs[1], 'H', axis_field=1)

plt.tight_layout()
# plt.savefig('slab.pdf', bbox_inches='tight')
```

## Conductor

```python
solver = slhfdtd.Solver((2e-6, 0, 0), grid_dist=1e-9)
solver.add_source(slhfdtd.PointSource((2e-9, 0, 0)))
solver.add_object(slhfdtd.Slab((1e-6, 0, 0), (1.3e-6, 0, 0),
                               permittivity=3, conductivity=2))
solver.add_boundary(slhfdtd.Exact1DAbsorber(0))
solver.run(2e-14)

fig, axs = plt.subplots(2)
visualizer = slhfdtd.Visualizer(solver)
visualizer.set_colors(object_colors='silver')
visualizer.plot1d_field(axs[0], 'E')
visualizer.plot1d_field(axs[1], 'H', axis_field=1)

plt.tight_layout()
# plt.savefig('conductor.pdf', bbox_inches='tight')
```

```python
solver = slhfdtd.Solver((2e-6, 0, 0), grid_dist=1e-9)
solver.add_source(slhfdtd.PointSource((2e-9, 0, 0)))
solver.add_object(slhfdtd.Slab((1e-6, 0, 0), (1.002e-6, 0, 0),
                               permittivity=3, conductivity=2))
solver.add_boundary(slhfdtd.Exact1DAbsorber(0))
solver.run(2e-14)

fig, axs = plt.subplots(2)
fig.suptitle(r'$x = 1.00\:\mu m$, $\mathrm{Thickness}'
             r' = 2.0\:\mathrm{nm}$, $\epsilon_r = 3.0$, '
             r'$\sigma = 2.0\:\mathrm{S/m}$')
visualizer = slhfdtd.Visualizer(solver)
visualizer.set_colors(object_colors='silver')
visualizer.plot1d_field(axs[0], 'E')
visualizer.plot1d_field(axs[1], 'H', axis_field=1)

plt.tight_layout()
# plt.savefig('nanolayer.pdf', bbox_inches='tight')
```
