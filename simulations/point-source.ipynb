# Point Sources

## Imports

```python
import slhfdtd
from matplotlib import pyplot as plt
```

## Global Configurations

```python
plt.rcParams.update({"text.usetex" : True})
```

## Point Current

```python
solver = slhfdtd.Solver((2e-6, 2e-6, 0))
solver.add_source(slhfdtd.PointSource((1e-6, 1e-6, 0), wavelength=200e-9))
solver.add_boundary(slhfdtd.AutoPML())
solver.run(1e-14)
```

```python
visualizer = slhfdtd.Visualizer(solver)
visualizer.set_aspect('equal')
visualizer.set_figsize((7.2, 5.4))
visualizer.set_interpolation_2d('none')

fig, axs = visualizer.plot2d()
# plt.savefig('point-current.pdf', bbox_inches='tight')
```

## Point Charge

```python
solver = slhfdtd.Solver((2e-6, 2e-6, 0))
solver.add_source(slhfdtd.PointSource((1e-6, 1e-6, 0),
                                      wavelength=200e-9,
                                      direction=None))
solver.add_boundary(slhfdtd.AutoPML())
solver.run(1e-14)
```

```python
visualizer = slhfdtd.Visualizer(solver)
visualizer.set_aspect('equal')
visualizer.set_figsize((7.2, 5.4))
visualizer.set_interpolation_2d('none')

fig, axs = visualizer.plot2d()
# plt.savefig('point-charge.pdf', bbox_inches='tight')
```
